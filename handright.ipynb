```python
# Preprocessing
# Let Separate Features and Target for machine Learning
# Step1 


features = list(mybigdata.columns[4:10])            # SepalLengthCm	SepalWidthCm	PetalLengthCm	PetalWidthCm	
target = mybigdata.columns[22]                      # Species

print('Features:',features)
print('Target:',target)

# store feature matrix in "X"
X = mybigdata.iloc[:,4:10]                          # slicing: all rows and 1 to 4 cols

# store response vector in "y"
y = mybigdata.iloc[:,22]                            # slicing: all rows and 5th col


print(y.shape)
print(X.shape)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(y)
y=le.transform(y)

# new col
mybigdata['EncodedState'] = y

print('States:',le.classes_)
print('Response variable after encoding:',y)
mybigdata.tail(10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X, y)
```

```python
#Step3: Prediction for new observation
value = knn.predict([[3,6,7,10,19,30]])
print('prediction value:',value)

print('Predicted State' , mybigdata.loc[mybigdata['EncodedState'] == 2, 'state'].values[0])
#data.loc[data['EncodedSpecies'] == 2, 'Species'].values[0]
```

```python
# more predictions for other rows

X_new = [[10, 20, 30, 40, 50, 60], [100, 110, 120, 70 ,80 ,90]]                        # Consider them as two new rows of input features in X
knn.predict(X_new)
```

```python
# Different value of K 
# instantiate the model (using the value K=5)

knn = KNeighborsClassifier(n_neighbors=5)

# fit the model with data
knn.fit(X, y)

# predict the response for new observations
print(knn.predict(X_new))

kypred = knn.predict(X)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y, kypred))
```
